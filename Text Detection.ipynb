```python
import cv2
import easyocr
import matplotlib.pyplot as plt
```

```python
# read image
image_path = "test1.png"

img = cv2.imread(image_path)
```

```python
# instance text detector
reader = easyocr.Reader(['en'], gpu=False)
```

```python
# detect text on image
text = reader.readtext(img)
```

```python
print(text)
```

```python
for t in text:
    print(t)
```

```python
# draw bbox and text
for t in text:
    print(t)
    
    bbox, text_, score = t
    
    cv2.rectangle(img, bbox[0], bbox[2], (0,255,0), 5)
    cv2.putText(img, text_, bbox[0], cv2.FONT_HERSHEY_COMPLEX, 0.65, (255,0,0), 2)
    
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
def read_text(image_path):
    img = cv2.imread(image_path)
    # instance text detector
    reader = easyocr.Reader(['en'], gpu=False)
    # detect text on image
    text = reader.readtext(img)
    # draw bbox and text
    for t in text:
        print(t)

        bbox, text_, score = t
        threshold = 0.25
        if score > threshold:
            cv2.rectangle(img, bbox[0], bbox[2], (0,255,0), 5)
            cv2.putText(img, text_, bbox[0], cv2.FONT_HERSHEY_COMPLEX, 0.65, (255,0,0), 2)

    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
read_text('test1.png')
```

```python
read_text('test2.png')
```

```python
read_text('test3.png')
```
